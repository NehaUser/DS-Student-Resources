# Merging data sets 

# Using pandas and merge function

```python
# Not specifying the common id

judgesAspectsUnique = pd.read_excel("judgesAspectsUnique.xlsx")
performances = pd.read_excel("performances.xlsx")

IceSkating = pd.merge(judgesAspectsUnique, performances)

IceSkating.head(10)
```

```python
# Specifying a unique common id using on=

IceSkating = pd.merge(judgesAspectsUnique, performances, on='performance_id')


IceSkating.head(10)
```

```python
# And if those keywords donâ€™t match up â€“ no need to sweat it! 
# Python also is set up to use common fields with different names, by specifying left_on= and right_on=:

IceSkating = pd.merge(judgesAspectsUnique, performances, left_on='performance_id', right_on='performance_id')

```

```python
performances.head()
```

```python
# What Happens When Columns are Named the Same, but Contain Different Data?

#Python will automatically add _x or _y to the column names to show from which dataset the data was derived.
```

# Specifying the Type of Join


```python
IceSkating = pd.merge(judgesAspectsUnique, performances, how='inner')
IceSkating = pd.merge(judgesAspectsUnique, performances, how='outer')
IceSkating = pd.merge(judgesAspectsUnique, performances, how='left')
IceSkating = pd.merge(judgesAspectsUnique, performances, how='right')
```

# Appending

# The way to append data in Python is by using the concatenate function in pandas: pd.concat.

```python
muscles1 = pd.read_excel("muscles1.xlsx")
```

```python
muscles2 = pd.read_excel("muscles2.xlsx")
```

```python
muscles3 = pd.concat([muscles1, muscles2])
```

```python
muscles3.head()
```

# You can get notified that the indices overlap if you add in the argument verify_integrity=True to your pd.concat() function:

```python
Muscles4 = pd.concat([muscles1, muscles2],verify_integrity=True)
```

# the easiest way to fix this is to just ignore the index, using the argument ignore_index=True, which will basically create a new one.

```python
Muscles5 = pd.concat([muscles1, muscles2], ignore_index=True)

# Instead of restarting the index
# again at zero once you switch to second dataset that was appended, it continues on to number 24.
```
