# Import pandas and the csv file

```python
import pandas as pd

airbnb_test_users = pd.read_csv("airbnb_test_users.csv")

airbnb_test_users.head(10)
```

# Q1. What is the average age of those who use each web browser type?


# Use aggregate function to calculate mean of age groupby first_browser

```python
airbnb_test_users.groupby('first_browser')['age'].mean()
```

# Q2. What is the total signup_flow for each device?

# Use aggregate function to calculate sum of signup_flow groupby first_device_type

```python
airbnb_test_users.groupby('first_device_type')['signup_flow'].sum()
```

# 1. They need the country information from the airbnb_sample_submission dataset included in the airbnb_test_users file.


```python
#import the airbnb_sample_submission csv file

airbnb_sample_submission = pd.read_csv("airbnb_sample_submission.csv")

airbnb_sample_submission.head()
```

# To include country information, we need to join/ merge the two data sets based on the id


```python
airnbnb_w_country = pd.merge(airbnb_test_users, airbnb_sample_submission, how='inner')
```

```python
airnbnb_w_country.head()
```

# 2. Add additional users to the test file from the airbnb_users dataset.

```python
#import the airbnb_users csv file

airbnb_users = pd.read_csv("airbnb_users.csv")
```

```python
# To add users from the above file, in the test file use concat function

airnbnb_users_country = pd.concat([airnbnb_w_country, airbnb_users])
```
