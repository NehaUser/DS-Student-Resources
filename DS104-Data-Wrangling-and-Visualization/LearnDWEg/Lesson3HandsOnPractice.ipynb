```python
import pandas as pd
```

```python
eating_habits = pd.read_csv("eating_habits.csv")
```

```python
eating_habits.head(10)

eating_habits.Activity.value_counts()
```

# Recode Activity into a new variable. Zeros should be not eating chocolate, and 1s should be eating chocolate.

```python
# Function to recode the activity 

def Activity_Rec(act):
    if act == "Eating candy, chocolate bars":
        return 1
    else:
        return 0

# Assign to a new variable
eating_habits["Activity1or0"] = eating_habits["Activity"].apply(Activity_Rec)

eating_habits.head(10)
```

# Recode Frequency from text to numbers in the same variable. The value zero should be the lowest frequency.

```python
eating_habits.Frequency.value_counts()
```

```python
cleanup_list = { "Frequency" : {"Never" : 0, "Seldom" : 1, "At least once a week" : 2, "Once a day" : 3, "More than once a day" : 4 }}
```

```python
eating_habits.replace(cleanup_list, inplace = True)
```

```python
eating_habits.head(10)
```

# Recode Sex from text to numbers in the same variable.

```python
cleanup_lst = {"Sex" : { "Males" : 0, "Females" : 1}}

eating_habits.replace(cleanup_lst, inplace = True)
```

# Dummy code the Age group variable.

```python
Age_grp = pd.get_dummies(eating_habits["Age group"])
```

```python
Age_grp.head()
```

```python
# To concat
eating_habits1 = pd.concat([eating_habits, Age_grp], axis=1)
eating_habits1.head()
```
