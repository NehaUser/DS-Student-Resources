```python
import pandas as pd
```

```python
PlaneFeatures = pd.read_excel("PlaneFeatures.xlsx")

PlaneFeatures.head(10)
```

```python
PlaneCandidates = pd.read_excel("PlaneCandidates.xlsx")

PlaneCandidates.head()
```

# Merge the above two data sets

```python
# Without using the common key

NewData = pd.merge(PlaneCandidates, PlaneFeatures)

NewData.head()
```

```python
# With specifying the common key - adshex

NewDatakey = pd.merge(PlaneCandidates, PlaneFeatures , on="adshex")

NewDatakey.head()
```

```python
# With outer join 

NewDataOuter = pd.merge(PlaneCandidates, PlaneFeatures, how='outer')

NewDataOuter.head()
```
