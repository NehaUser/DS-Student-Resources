```python
babies['Footprint'] = 'Y'
```

```python
babies.head()

```

```python
babies.rename(columns = {"ParentPhoneNumber" : "Phone"}, inplace = True)
```

```python
# To combine data of two or more columns together

# Adds a new column Address in the same dataset
# map(str) coverts it into string if not already
babies["Address"] = babies["StreetAddress"] + "/" + babies["City"] + "/ " + babies["Zipcode"].map(str)
```

```python
# To break columns into different columns

#The expand=True argument is very important – 
#it makes each separated section its own column in a data frame rather than just producing a list.

babies1 = babies['Address'].str.split('/', expand=True)

babies1.head()


```

```python
#OR 
#To rename the columns of the dataset as well

babies2 = babies['Address'].str.split('/', expand = True).rename(columns = lambda x: "Address" + str(x+1))

babies2.head()
```

```python
# The next step is to add those columns back into your dataset
# you will append your data side by side.
# In Python, the pandas package has a function called .concat() that will add in columns or rows.
# specify axis=1 to tell Python that you are adding columns, not rows.

babies3 = pd.concat([babies, babies2], axis = 1)

babies3.head()
```

```python
# Subsetting Rows
babies4 = babies[:10] # it will only take first 3 rows with index being 2 (0,1,2)
babies4.head() 

babies3.head()

# Subsetting Columns

babies5 = babies4[['Name', 'Birthday']]
babies5.head()
```
