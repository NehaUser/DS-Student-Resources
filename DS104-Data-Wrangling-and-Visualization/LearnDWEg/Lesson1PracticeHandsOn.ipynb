# Separate the url column out so that you can see in one column the website and in the second column the domain.

```python
FakeNews_Split = FakeNews['url'].str.split('_', expand =True).rename(columns = lambda x: 'Col' + str(x+1))
```

```python
# Add the split dataset back in FakeNews dataset using concat function

FakeNews_C = pd.concat([FakeNews , FakeNews_Split], axis = 1)
```

# Unite the Domain back

```python
FakeNews_C['urlnew'] = FakeNews_C['Col1'] + '_' + FakeNews_C['Col2'].map(str)

FakeNews_C.drop(['url'], axis = 1)
```
