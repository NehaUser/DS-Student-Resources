# Correlation activity between different variables of a table

## Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##  Load data

```python
power_lift = pd.read_csv("power_lifting.csv")
power_lift.head()
```

## Remove all categorical variables/ not-continuous varibales for correlation


```python
power_lift1 = power_lift.drop(["Name", "Sex", "Equipment", "Division"], axis=1)
```

```python
power_lift1.dropna()
```

```python
list(power_lift1.columns)
```

## Individual Correlation between Bodyweight and BestBench


```python
power_lift1['BodyweightKg'].corr(power_lift1['BestBenchKg'])
```

```python
# It means body weight and best bench press are positively correlated
```

## Correlation for the whole data by creating  correlation matrix

```python
power_lift1.corr(method = 'pearson')
```

### Graphic presentation using matplotlib

```python
power_lift1.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# Darker red and pink shades shows strong positive correlation
# Darker blue and blue shades show strong weak negative correlation
```

### Visual using sns heatmap

```python
sns.heatmap(power_lift1.corr(), annot=True)
```

```python
# We can see best squat is strongly positively correlated with total weight.
# Also body weight and weight class are strongly positively correlated to each other
```

```python
# Interpretation / Discussion
# For either chart, larger correlations are shown in warmer colors.
# Ignore the diagonal line of dark color down the center, since this is just all the variables correlating
# with themselves. Then focus on the darker ones and move down. Looks like both body weight and dead
# lifting weight correlate pretty highly with a number of other factors.

# Body weight is correlated with:

# Best squat weight
# Best bench weight
# Best deadlift weight

# Best dead lift weight is correlated with:

# Best squat weight
# Best bench weight
# Total kilograms
# Wilks

# In addition, best squat is correlated with:

# Best bench weight
# Total kilograms
# Wilks
```
